# Box Plot

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70]

plt.boxplot(x)

plt.show()
```

```python
x=[10,20,30,40,50,60,70]

plt.boxplot(x,notch=True)

plt.show()
```

```python
x=[10,20,30,40,50,60,70]

plt.boxplot(x,vert=False)

plt.show()
```

```python
x=[10,20,30,40,50,60,70]

plt.boxplot(x,widths=0.3)

plt.show()
```

```python
x=[10,20,30,40,50,60,70,120]

plt.boxplot(x,labels=["python"],
            patch_artist=True,showmeans=True,sym="g+",boxprops=dict(color="r"),
           capprops=dict(color="m"),whiskerprops=dict(color="r"),
            flierprops=dict(markeredgecolor="y"))

plt.show()
```

```python
x=[10,20,30,40,50,60,70,120]

plt.boxplot(x,labels=["python"],
            patch_artist=True,showmeans=True,whis=2)

plt.show()
```

```python
x=[10,20,30,40,50,60,70,120]
x1=[10,20,40,50,30,0,90]

y=[x,x1]

plt.boxplot(y,labels=["python","c++"],
            showmeans=True,sym="g+",boxprops=dict(color="r"),
           capprops=dict(color="m"),whiskerprops=dict(color="r"))


plt.show()
```
