# Steam Plot

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6]
y = [2,2,5,7,3,8]

plt.stem(x,y)

plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [2,2,5,7,3,8]

plt.stem(x,y,linefmt=":",markerfmt="+",bottom=0,basefmt="g",label="python",use_line_collection=False,
        orientation="horizontal")

plt.legend()
plt.show()
```
