# Scatter Plot


```python
import matplotlib.pyplot as plt

day= [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
plt.scatter(day,no,color="r")
plt.title("scatter plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No",fontsize=15)
plt.show()
```

```python
day= [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
colors = ["r","y","g","b","r","g","r"]
sizes=[50,20,4,5,2,10,7]
plt.scatter(day,no,c=colors,s=sizes,alpha=0.5,marker="*",edgecolor="g",linewidth=2) 
# alpha ni value flot(0.1)ma hoy 
plt.title("scatter plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No",fontsize=15)
plt.show()
```

```python
day= [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
colors = [10,49,30,29,56,20,30]
sizes=[50,20,40,50,20,10,70]
plt.scatter(day,no,c=colors,s=sizes,cmap="viridis",alpha=0.5) 
t=plt.colorbar()
t.set_label("ColorBar",fontsize=15)
plt.title("scatter plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No",fontsize=15)
plt.show()
```

```python
day= [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
no2 = [3,2,4,5,1,6,2]
colors = [10,49,30,29,56,20,30]
sizes=[50,20,40,50,20,10,70]
plt.scatter(day,no,c=colors,s=sizes,cmap="viridis",alpha=0.5) 
plt.scatter(day,no2,color="r",s=sizes,alpha=0.5) 
t=plt.colorbar()
t.set_label("ColorBar",fontsize=15)
plt.title("scatter plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No",fontsize=15)
plt.show()
```
