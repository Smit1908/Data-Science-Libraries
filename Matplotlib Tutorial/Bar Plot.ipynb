# Bar Plot

```python
import matplotlib.pyplot as plt

x = ["python","c","c++","java"]
y = [85,70,60,82]
plt.xlabel("language",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.title("wscube",fontsize=15)
c = ["r","y","g","b"]
plt.bar(x,y,width=0.2,color=c,align="center",edgecolor="m",linewidth=3,linestyle=":",alpha=0.5)
# candle ni width like jadi,patali karva mate
# align niche na name ni palce change karva mate like that right center left
plt.show()
```

```python
x = ["python","c","c++","java"]
y = [85,70,60,82]
z = [20,30,40,50]
plt.xlabel("language",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.title("wscube",fontsize=15)
plt.bar(x,y,width=0.2,color="r",label="popularity")
plt.bar(x,z,width=0.2,color="y",label="popularity1")
plt.legend()
plt.show()
        
```

```python
import numpy as np
x = ["python","c","c++","java"]
y = [85,70,60,82]
z = [20,30,40,50]

width = 0.2

p = np.arange(len(x))
p1 = [j+width for j in p]
plt.xlabel("language",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.title("wscube",fontsize=15)

plt.bar(p,y,width,color="r",label="popularity")
plt.bar(p1,z,width,color="y",label="popularity1")

plt.xticks(p+width/2,x,rotation=20)# nichena fount maate x pr chhe te
plt.legend()
plt.show()
```

```python
x = ["python","c","c++","java"]
y = [85,70,60,82]
z = [20,30,40,50]

plt.xlabel("language",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.title("wscube",fontsize=15)

plt.barh(x,y,color="r",label="popularity")
plt.barh(x,z,color="y",label="popularity1")

plt.legend()
plt.show()
```
