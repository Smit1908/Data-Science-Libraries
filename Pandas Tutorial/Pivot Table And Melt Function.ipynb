# Melt Function

```python
import pandas as pd

var = pd.DataFrame({"days":[1,2,3,4,5,6],"eng":[10,12,14,15,16,12],"maths":[17,18,19,13,14,16]})

var
```

```python
pd.melt(var,id_vars=["days"],var_name="python",value_name="wscube")
```

# Pivot Table

```python
var1 = pd.DataFrame({"days":[1,2,3,4,5,6],
                     "st_name":["a","b","c","a","b","c"],"eng":[10,12,14,15,16,12],"maths":[17,18,19,13,14,16]})

var1
```

```python
var1.pivot(index="days",columns="st_name",values="eng")
```

```python
var1.pivot(index="days",columns="st_name")
```

```python
var2 = pd.DataFrame({"days":[1,1,1,1,2,2],
                    "st_name":["a","b","a","b","a","b"],"eng":[10,12,14,15,16,12],"maths":[17,18,19,13,14,16]})

var2
```

```python
var2.pivot_table(index="st_name",columns="days",
                aggfunc="mean",margins="True")
```

```python
var2.pivot_table(index="st_name",columns="days",
                aggfunc="sum",margins="True")
```
