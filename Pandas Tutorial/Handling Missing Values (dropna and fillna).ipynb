# Dropna & Fillna

```python
import pandas as pd

var = pd.read_csv("C:\\Users\\Admin\\python\\Pandas tutorials\\Test_4.csv")

var
```

```python
var.dropna(axis=1) 

# var.dropna()

# nun value delete karva mate 

# aapda excel ma nun value nathi so samajava mate j chhe function

# samjava mate Wscube tech na pandas playlist ma joi levu
```

# Fillna

```python
var.fillna("pyhton")

# nun value pr aapeli value fill kari de che
```
