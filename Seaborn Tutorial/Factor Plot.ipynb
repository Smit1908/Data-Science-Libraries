# Factor Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt 
import pandas as pd

var = sns.load_dataset("tips")
var
```

```python
sns.factorplot(x="size",y="tip",data=var,hue="sex")
plt.show()
```

```python
sns.factorplot(x="size",y="tip",data=var,hue="sex",kind="box")
plt.show()
```
