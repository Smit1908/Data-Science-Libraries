# Violin Plot

```python
import seaborn as sns 
import matplotlib.pyplot as plt
import pandas as pd

```

```python
var = sns.load_dataset("tips")
var
```

```python
sns.violinplot(x="day",y="total_bill",data = var,hue="time",color="r")
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data = var,linewidth=3,palette="twilight_r",
              saturation=0.4)
plt.show()
```

```python
sns.violinplot(x="time",y="total_bill",data = var,order=["Dinner","Lunch"],
               inner="quart")
#inner parameter ={"box","quartile","point","stick",None}
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data = var,hue="sex",split=True,
               scale="width")
plt.show()
```

```python
sns.violinplot(x=var["total_bill"])
plt.show()
```
