# Strip Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

var = sns.load_dataset("tips")
var
```

```python
m = {"male":"o","female":"<"}
sns.stripplot(x="day",y="total_bill",data=var,hue="sex",palette="turbo",
              linewidth=1,edgecolor="r",size=10,marker="<",alpha=0.2)
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",data=var,hue="sex",palette="turbo",
              linewidth=1,edgecolor="r",jitter=5)
plt.show()
```

```python
sns.stripplot(y=var["total_bill"]) # horizontal mate y and vertical mate x
plt.show()
```
