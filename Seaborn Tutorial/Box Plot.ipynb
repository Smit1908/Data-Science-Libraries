# Box Plot 

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

var = sns.load_dataset("tips")
var
```

```python
sns.set(style="whitegrid")

sns.boxplot(x="day",y="total_bill",data=var,hue="sex",color="r",
            order=["Fri","Sun","Thur","Sat"],showmeans=True,
           meanprops={"marker":"+","markeredgecolor":"y"},
           linewidth=3,palette="cubehelix_r")

plt.show()
```

```python
sns.set(style="whitegrid")

sns.boxplot(x="total_bill",y="day",data=var,color="r",
            order=["Fri","Sun","Thur","Sat"],showmeans=True,
           meanprops={"marker":"+","markeredgecolor":"y"},
           linewidth=3,palette="cubehelix_r")

plt.show()
```

```python
sns.set(style="whitegrid")

sns.boxplot(data=var,color="r",showmeans=True,
           meanprops={"marker":"+","markeredgecolor":"y"},
           linewidth=3,palette="cubehelix_r",orient="h")

plt.show()
```

```python
sns.set(style="whitegrid")

sns.boxplot(x=var["total_bill"])

plt.show()
```
