# Bar Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

var = sns.load_dataset("penguins")
var
```

```python
order_1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=var,color="m")
plt.show()
```

```python
sns.set(style="darkgrid")
order_1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=var,hue="sex",
           order = order_1,hue_order=["Female","Male"],ci=100,n_boot=2,
           palette="icefire",errcolor="r",errwidth=2,capsize=0.2,dodge=True,
           alpha=0.7)

# (saturation=4) this is a parameter for bluring 

plt.show()
```

```python
order_1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x="bill_depth_mm",y="bill_length_mm",data=var,orient="v")
plt.show()
```
