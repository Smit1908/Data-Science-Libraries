# Heatmap Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np



```

```python
var = np.linspace(1,10,20).reshape(4,5)
var
```

```python
y={"fontsize":20,"color":"r"}
v = sns.heatmap(var,vmin=0,vmax=10,cmap="PuOr",annot=True,annot_kws=y,linewidth=7,
           linecolor="g",cbar=False,xticklabels=False,yticklabels=False)
v.set(xlabel="python",ylabel="wscube")
sns.set(font_scale=1)
plt.show()
```

```python
data = sns.load_dataset("anagrams")
x = data.drop(columns=["attnr"],axis=1).head(10)
x
```

```python
sns.heatmap(x)
plt.show()
```

```python
var1 = np.linspace(1,10,10).reshape(2,5)
var1
```

```python
ar = np.array([["a0","a1","a2","a3","a4"],
               ["b0","b1","b2","b3","b4"]])
ar
```

```python
sns.heatmap(var1,vmin=0,vmax=10,cmap="PuOr",annot=ar,fmt="s")
plt.show()
```
