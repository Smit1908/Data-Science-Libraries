# Scatter Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var = sns.load_dataset("penguins")
var
```

```python
m = {"Male":"*","Female":"o"}
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=var,hue="sex",
                style="sex",size="sex",sizes=(100,60),palette="YlOrRd",
               alpha=0.8,markers=m)


plt.show()
```
