# Pair Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

var = sns.load_dataset("tips")
var
```

```python
sns.pairplot(var,vars=["total_bill","tip"],hue="sex",
             hue_order=["Female","Male"])
plt.show()
```

```python
sns.pairplot(var,hue="sex",hue_order=["Female","Male"],x_vars=["total_bill","tip"])
plt.show()
```

```python
sns.pairplot(var,hue="sex",hue_order=["Female","Male"],kind="reg",diag_kind="hist")
#kind parameter : "scotter","kde","hist","reg"
plt.show()
```

```python
sns.pairplot(var,hue="sex",hue_order=["Female","Male"],markers=["*","<"])
plt.show()
```

```python
sns.pairplot(var)
plt.show()
```
